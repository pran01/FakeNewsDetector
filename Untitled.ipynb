```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import re
import numpy as np
#Anaconda Navigator: conda install -c conda-forge textblob
from textblob import TextBlob
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#converting csv files to pandas object(dataframe)
real_df=pd.read_csv('True.csv')
fake_df=pd.read_csv('Fake.csv')
```

```python
real_df.drop('text',axis=1,inplace=True)
fake_df.drop('text',axis=1,inplace=True)
```

```python
real_df['Authenticity']=['Real' for i in range(real_df.shape[0])]
fake_df['Authenticity']=['Fake' for i in range(fake_df.shape[0])]
```

```python
real_df.drop('date',axis=1,inplace=True)
fake_df.drop('date',axis=1,inplace=True)
```

```python
real_and_fake=pd.concat([real_df,fake_df],ignore_index=True)
```

```python
real_and_fake['title_length'] = real_and_fake['title'].str.count(' ') + 1
```

```python
#preprocessing/cleaning
#removing duplicates
real_and_fake=real_and_fake.drop_duplicates()
#making text lower
real_and_fake['title']=real_and_fake['title'].str.lower()
#replacing realwith 0 and fake  with 1
real_and_fake['Authenticity']=real_and_fake['Authenticity'].replace({'Real':0,'Fake':1})
#removing text in brackets
real_and_fake['title'].replace(to_replace='\(.*?\)', value='', regex=True,inplace=True)
#removing punctutatuations and symbols
real_and_fake['title'].replace(to_replace='[.,:\–\-\'\"\#\?$;]', value='', regex=True,inplace=True)
real_and_fake['title'].replace(to_replace='[‘’“”…]', value='', regex=True,inplace=True)
#subject is not making much sense either
real_and_fake.drop('subject',axis=1,inplace=True)
```

```python
#Lets find out how much bad words are contained in title
#its clear by looking at dataset that if the title contains *, it contains profanity
real_and_fake['profanity'] = real_and_fake['title'].str.count('\*')
#counting * in title is not the same as counting bad words, since some bad words contain multiple *
```

```python
#lets do sentiment analysis now.
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity
real_and_fake['polarity']=real_and_fake['title'].apply(pol)
real_and_fake['subjectivity']=real_and_fake['title'].apply(sub)
```

```python
#EDA(Exploratory Data Analysis)
plt.plot(real_and_fake[real_and_fake['Authenticity']==0].title_length,color='red')
plt.plot(real_and_fake[real_and_fake['Authenticity']==1].title_length,color='yellow')
plt.show()
#this shows that the variation of title-length is more in fake news which ranges from 1 to 45.
plt.plot(real_and_fake[real_and_fake['Authenticity']==0].profanity,color='red')
plt.plot(real_and_fake[real_and_fake['Authenticity']==1].profanity,color='yellow')
plt.show()
#this one was clear from the start, only fake news contain profanity.
plt.plot(real_and_fake[real_and_fake['Authenticity']==0].polarity,color='red')
plt.plot(real_and_fake[real_and_fake['Authenticity']==1].polarity,color='yellow')
plt.show()
#polarity is not helping much
plt.plot(real_and_fake[real_and_fake['Authenticity']==0].subjectivity,color='red')
plt.plot(real_and_fake[real_and_fake['Authenticity']==1].subjectivity,color='yellow')
plt.show()
#subjectivity seems even worse as a feature, it just looks like two huge rectangles xDD
#lets remove subjectivity then
real_and_fake.drop('subjectivity',axis=1,inplace=True)
```

```python
#The data looks good now, let's pickle this dataset
real_and_fake_ml=real_and_fake.drop('title',axis=1)
real_and_fake_ml.to_pickle("real_and_fake_ml.pkl")
real_and_fake_dl=real_and_fake[['title','Authenticity']]
real_and_fake_dl.to_pickle("real_and_fake_dl.pkl")
```

```python
#lets try ML first.
real_and_fake_ml=pd.read_pickle("real_and_fake_ml.pkl")
#feature selection
feature_names=np.array(real_and_fake_ml.columns[1:5])
target=np.array(real_and_fake_ml['Authenticity'])
features=np.array(real_and_fake_ml[feature_names[:]])
#normalising the features is best for some models
scaler=StandardScaler()
features=scaler.fit_transform(features)
```

```python
#svm.svc(Support Vector Machine)
from sklearn import svm
#svc_clf1=svm.SVC(kernel='linear',C=1)
#svc_clf1.fit(features,target)
#print(cross_val_score(svc_clf1,features,target,cv=10).mean())
#svc_clf2=svm.SVC(kernel='poly',degree=3,C=1,gamma='scale')
#svc_clf2.fit(features,target)
#print(cross_val_score(svc_clf2,features,target,cv=10).mean())
#svc_clf3=svm.SVC(kernel="sigmoid",C=1,gamma='scale')
#svc_clf3.fit(features,target)
#print(cross_val_score(svc_clf3,features,target,cv=10).mean())
svc_clf4=svm.SVC(kernel="rbf",C=1,gamma='scale')
svc_clf4.fit(features,target)
print(cross_val_score(svc_clf4,features,target,cv=10).mean())
#rbf kerel worked best (84.24%)
```

```python
#XGBoost(Extreme Gradient Boosting)
import xgboost as xgb
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=0)
#xgboost expects the data in Dmatrix
train = xgb.DMatrix(X_train, label=y_train)
test = xgb.DMatrix(X_test, label=y_test)
param={
    'max_depth':4,
    'eta': 0.6,
    'objective': 'multi:softmax',
    'num_class':2
}
epochs=25
model = xgb.train(param, train, epochs)
predictions = model.predict(test)
print(accuracy_score(y_test, predictions))
#84.74%. XGBoost works better than svm with rbf kernel.
```

```python
#multinomial Naive Bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
featuresminmax=scaler.fit_transform(features)
mnb_clf=MultinomialNB()
mnb_clf.fit(featuresminmax,target)
cross_val_score(mnb_clf,featuresminmax,target,cv=10).mean()
```
